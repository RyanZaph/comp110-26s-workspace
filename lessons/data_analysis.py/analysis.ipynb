## `read_csv_rows'

- Function name: read_csv_rows
- Parameter: `str' path to csv file
- Return type: `list[dict{str,str}]

This function reads csv files annd gives me a data structure to work with

```python
def read_csv_rows(filename: str) -> list[dict[str,str]]:
    """Read the rows of a CSV"""
    result: list[dict[str,str]]=[]

    # Open the file to read it
    file_handle = open(filename, "r", encoding="utf8")

    # Make a "DictReader"
    csv_reader = DictReader(file_handle)

    # Copy each "row" of csv reader into our list
    for row in csv_reader:


            # Copy each "row" of csv reader into our list
        result.append(row)

            # Close the file 
    file_handle.close()

    return result


    
```

```python
read_csv_rows("walter_wally.csv")
```
